```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2
#anropa funktionen med parametern 5
print(f(5))
x = np.linspace(-10,10)
y = f(x)
print(y)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(x)=x**2")
plt.show()
plt.grid()
```
