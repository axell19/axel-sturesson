```python
# d)
# Här printade jag 10 random punkter mellan -1 och 1 genom att använda random.uniform modulen. Jag gjorde också så att talen skulle avrundas till 2 decimaler. 
import random as rnd
for i in range(0,11):
    x=round(rnd.uniform(1,-1),2)
    y=round(rnd.uniform(1,-1),2)
    print(f"punkt{i}:({x},{y})")
```

```python
# e)

# Här gjorde jag så att användaren får välja antalet punket som ska printas och hur stor radien ska vara. 
punkter=int(input("Skriv in antalet punkter du vill ha"))
radie=float(input("Mata in hur stor cirkelns radie ska vara"))
import matplotlib.pyplot as plt
import random as random
antalpunkter=0
innanför=0
import random as rnd
# Här gjorde jag en for sats som printar ut random tal(punkter) mellan -1 och 1 vilket är x och y punkterna
for i in range(0,punkter):
    x=round(rnd.uniform(1,-1),5)
    y=round(rnd.uniform(1,-1),5)
    antalpunkter +=1
# Här använder jag pythagoras sats för att kunna räkna ut anvståndet mellan origo och punkten. 
    if x**2 + y**2 < radie:
        innanför +=1
        plt.plot(x,y,'-o')

print(f"{innanför} av {antalpunkter} punkter hamnar inuti cirkeln vilket är {innanför/antalpunkter*100}%")
```

```python
# d)
# Här gjorde jag så att användaren får mata in två olika tal mellan 1 och 100. 
tal = int(input("mata in ett tal mellan 1 och 100"))
tal2 = int(input("mata in ett till tal mellan 1 och 100"))
# Och här har jag gjort likdant (om talet är delbart på det talet användaren skrev in ska "burr" printas). Dock så la jag också till så att om talet är delbart med det andra talet användaren skrev in ska "birr" printas. 
for i in range(1, 101):
    if i%tal==0:
        print("Burr", end=" ")
    elif i%tal2==0:
        print("Birr",end=" ")
    else:
        print(i,end=" ")


    

```

```python
# e)
# Här får användaren välja en egen slut och start multipel och så får den ange två egna tal. 
start=int(input("mata in en startmultipel"))
slut=int(input("mata in en slutmultipel"))
tal = int(input("mata in ett tal"))
tal2 = int(input("mata in ett till tal"))
# slut+=1 skrev jag för att i python så minsakr den alltid talet man skriver in i en for sats med 1, tex om man vill ha 1 till 100 så får man skriva 1 och 101 men det behövs inte nu för att jag ka till slut+=1.
slut+=1
# Och här gjorde jag samma som förra uppgiften
for i in range(start, slut):
    if i%tal==0:
        print("Burr", end=" ")
    elif i%tal2==0:
        print("Birr",end=" ")
    else:
        print(i,end=" ")
```
